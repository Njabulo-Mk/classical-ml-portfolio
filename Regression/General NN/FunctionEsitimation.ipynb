```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRFRegressor
from sklearn.neural_network import MLPRegressor

from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
X = np.linspace(-2*np.pi, 2*np.pi, 1000).reshape(-1,1)
y = np.sin(X).ravel()
```

```python
split_idx = int(0.8 * len(X))
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]
```

```python
LR = LinearRegression()
LR.fit(X_train, y_train)
y_pred_LR = LR.predict(X_test)
```

```python
sort_idx = X_test[:,0].argsort()
X_test_sorted = X_test[sort_idx]
y_test_sorted = y_test[sort_idx]
y_pred_LR_sorted = y_pred_LR[sort_idx]

plt.figure(figsize=(12,6))
plt.plot(X_test_sorted, y_test_sorted, color='blue', label='True sin(x)')
plt.plot(X_test_sorted, y_pred_LR_sorted, color='red', label='Linear Regression Prediction')
plt.legend()
plt.title("Function Approximation: Linear Regression vs True sin(x)")
plt.show()
```

```python
RF = RandomForestRegressor(n_estimators=200, min_samples_leaf=1, random_state=42)
RF.fit(X_train, y_train)
y_pred_RF = RF.predict(X_test)
```

```python
sort_idx = X_test[:,0].argsort()
X_test_sorted = X_test[sort_idx]
y_test_sorted = y_test[sort_idx]
y_pred_RF_sorted = y_pred_RF[sort_idx]

plt.figure(figsize=(12,6))
plt.plot(X_test_sorted, y_test_sorted, color='blue', label='True sin(x)')
plt.plot(X_test_sorted, y_pred_RF_sorted, color='red', label='Random Forest Prediction')
plt.legend()
plt.title("Function Approximation: Random Forest Regression vs True sin(x)")
plt.show()
```
